```python
import torch

import pandas as pd
import numpy as np
from scipy.special import softmax

from time import sleep
import json
import os

from collections import namedtuple

from tqdm.notebook import tqdm

from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
model_type = f"cardiffnlp/twitter-roberta-base-sentiment-latest"

tokenizer = AutoTokenizer.from_pretrained(model_type)
model = AutoModelForSequenceClassification.from_pretrained(model_type)
```

```python
def all_sentiment_scores(tweet):
    """To return the sentiment score of a Tweet as analysed by BERTweet. """
    tokens = tokenizer.encode(tweet, return_tensors='pt')
    result = model(tokens)
    return softmax(result.logits.detach().numpy())
```

```python
def max_sentiment_score(tweet):
    """To return the sentiment score of a Tweet as analysed by BERTweet. """
    return np.argmax(all_sentiment_scores(tweet)) - 1
```

```python
df=pd.read_csv('/content/tweetdataset.csv',on_bad_lines='skip',delimiter='\t')
df.columns=['Category','Date','Tweet']
df.head()
```

```python
df['Score']=df['Tweet'].apply(all_sentiment_scores)
df.head()
```

```python
df['neg_score']=''
df['neu_score']=''
df['pos_score']=''
for i in range(len(df['Score'])):
  df['neg_score'][i]=df['Score'][i][0][0]
  df['neu_score'][i]=df['Score'][i][0][1]
  df['pos_score'][i]=df['Score'][i][0][2]
df.head()
```

```python
df['MaxScore']=df['Tweet'].apply(max_sentiment_score)
df.head()
```

```python
agg_df = df.groupby(['Category', 'Date']).agg({'neg_score': 'sum','neu_score': 'sum','pos_score': 'sum'})
agg_df
```

```python
agg_df1 = df.groupby(['Category', 'Date', 'MaxScore'])['MaxScore'].count().reset_index(name='Count')
pivoted_df = agg_df1.pivot_table(index=['Category', 'Date'], columns='MaxScore', values='Count', fill_value=0)
pivoted_df = pivoted_df.reset_index()
pivoted_df.columns=['Category','Date','Neg_count','Neu_count','Pos_count']
pivoted_df
```

```python
finaldf=pd.merge(pivoted_df,agg_df,left_on='Category',right_on='Category')
final_df = finaldf.groupby(['Category', 'Date']).agg({'neg_score': 'sum','neu_score': 'sum','pos_score': 'sum','Neg_count': 'sum','Neu_count': 'sum','Pos_count': 'sum'})
final_df
```

```python
for i in range(len(finaldf['Score'])):
  df['wt_neg'][i]=df['Score'][i][0][0]*df['Count']
  df['wt_neu'][i]=df['Score'][i][0][1]*df['Count']
  df['wt_pos'][i]=df['Score'][i][0][2]*df['Count']
df.head()
```
